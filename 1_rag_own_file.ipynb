```python
from langchain_core.documents import Document

my_text="""Ramesh is a very poor guy who lives in a small village name is Arthe Bk. He is the only son of his parents. His mother name is Pramila and father name is Onkar. He also has a younger sister name is shivani who is married and lives in Nashik city. He also has a elder brother name is Pravin who is working in Nashik city he has a small business of fruits. One day Ramesh decided to go to Pune city to find a job. He packed his bags and started his joureny to Pune via Mahalaxmi Train. On the way to Pune he met with an accident lost his money. Now ramesh is stuck in Pune city with no money and no contacts. He was very hungry and thirsty so he decided to go to a nearby restaurant to have some food. But since he does not have any money he was unable to pay for the food. So he decided to work in the restaurant to pay for his food. Restaurant name is Fish on Dish"""

docs=[Document(page_content=my_text, metadata={"source":"Marathi story","documentId":"Doc1"})]
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=220,
    chunk_overlap=50)

chunks = splitter.split_documents(docs)
chunks
```

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embedding_model
)
```

```python
vectorstore.similarity_search("city?", k= 1)
```
