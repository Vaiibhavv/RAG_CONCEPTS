```python
import os
from dotenv import load_dotenv
load_dotenv()

if os.environ["OPENAI_API_KEY"]:
    print("API Key is set")
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-5-nano",temperature=0)
```

### **STEP1: IMPORT THE PDF FILE AS TEXT , extracting text from pdf** ##

```python
from langchain_community.document_loaders import PyPDFLoader
file_path="Docs\science_notes.pdf" 
loader=PyPDFLoader(file_path)
docs =loader.load()
docs
```

```python
docs[0].metadata
```

#### Creating own metadata for pdf chunks ####

```python
for i in docs:
    i.metadata={"source":"Science Book", "Author":"NCERT"}
```

### **STEP2: SPLITTING THE PDF INTO DOC CHUNKS** ##

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=600, chunk_overlap=100)
chunks=text_splitter.split_documents(docs)
chunks
```

```python
len(chunks)
```

### **STEP3: CONVERT DOCS TO EMBEDDING** ##

```python
from langchain_openai import OpenAIEmbeddings
embeddings_model=OpenAIEmbeddings(model="text-embedding-3-small")
```

### **STEP4: STORE EMBEDDINGS TO VECTORE STORE** ##

```python
from langchain_community.vectorstores import Chroma

vectorstore=Chroma.from_documents(
    documents=chunks,
    embedding=embeddings_model
)
```

### **STEP5: SEMANTIC SEARCH** ##

```python
context=vectorstore.similarity_search("what is energy",k=1)
```

```python
response=llm.invoke(f'what is energy? based on the context: {context}')
print(response.content)
```
